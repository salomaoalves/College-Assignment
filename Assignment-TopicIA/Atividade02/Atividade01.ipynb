```
#CELULA 01

import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive/')
```

```
#CELULA 02

# ler o arquivo csv
df = pd.read_csv('/content/gdrive/My Drive/Agrupamentos/file.csv', header=None)
df.head()
```

```
#CELULA 03

n_row = df.shape[0] # numero de linhas
n_col = df.shape[1] # numero de colunas
col = df.columns # listas cm o nome das colunas
```

```
#CELULA 04

col_nan = [] # lista da colunas que tem valor null

# pega as colunas que possuem algum valor null
for i in range(n_row):
  for j in range(n_col):
    if df.iloc[i,j] == '?' and col[j] not in col_nan:
      col_nan.append(col[j])
```

```
#CELULA 05

moda = {} # moda (value) final de cada coluna (key)

# calcula a moda de cada coluna, uma por vez
for col in col_nan:
  moda_col = {} # dicionario com cada valor (key) e a qtd de vezes q ele aparece (value)
  for i in range(n_row):

    if df.iloc[i,col] != '?':
      if df.iloc[i,col] not in moda_col.keys():
        moda_col[df.iloc[i,col]] = 1
      else:
        moda_col[df.iloc[i,col]] += 1

  moda[col] = max(moda_col, key=moda_col.get) # salva a moda da coluna atual
```

```
#CELULA 06

# substitui os valor nulos pela moda
for col in moda.keys():
  df[col][df[col] == '?'] = moda[col]
```

```
#CELULA 07

# gera o arquivo csv como resposta
df.to_csv('/content/gdrive/My Drive/Agrupamentos/update_file.csv')
```
